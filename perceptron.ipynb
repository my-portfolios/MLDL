```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#traning data generate
m = 100
x1 = 8*np.random.rand(m, 1)
x2 = 7*np.random.rand(m, 1) - 4

g = 0.8*x1 + x2 - 3
```

```python
C1 = np.where(g >= 1)
C2 = np.where(g < -1)
```

```python
X1 = np.vstack([np.ones([C1[0].shape[0],]), x1[C1], x2[C1]]).T
X2 = np.vstack([np.ones([C2[0].shape[0],]), x1[C2], x2[C2]]).T
X = np.vstack([X1, X2])

y = np.vstack([np.ones([C1[0].shape[0], 1]), -np.ones([C2[0].shape[0], 1])])

X = np.asmatrix(X)
y = np.asmatrix(y)
```

```python
w = np.ones([3, 1])
w = np.asmatrix(w)

n_iter = y.shape[0]
flag = 0

while flag == 0:
  flag = 1
  for i in range(n_iter):
    if y[i, 0] != np.sign(X[i,:]*w)[0,0]:
      w += y[i,0]*X[i,:].T
      flag = 0
```

```python
x1p = np.linspace(0,8,100).reshape(-1,1)
x2p = - w[1,0]/w[2,0]*x1p - w[0,0]/w[2,0]

plt.figure(figsize=(10, 8))
plt.plot(x1[C2], x2[C2], 'ro', alpha = 0.4, label = 'C2')
plt.plot(x1[C1], x2[C1], 'bo', alpha = 0.4, label = 'C1')
plt.plot(x1p, x2p, c = 'k', linewidth = 3, label = 'perceptron')
plt.xlim([0, 8])
plt.xlabel('$x_1$', fontsize = 15)
plt.ylabel('$x_2$', fontsize = 15)
plt.legend(loc = 1, fontsize = 15)
plt.show()
```
